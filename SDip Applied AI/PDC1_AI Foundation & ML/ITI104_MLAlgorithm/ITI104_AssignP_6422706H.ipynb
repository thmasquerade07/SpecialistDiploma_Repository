##### 2.Split Data - Train(70) | Test(30)

```python
from sklearn.model_selection import train_test_split

# Split dataset to train and test sets
X = df[iris.feature_names].values  # Extract feature values
y = df['Species'].values  # Extract target values
print('Data size', X.shape)
print('Label size', y.shape)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, shuffle=True)
print('Train data size', y_train.shape[0])
print('Test data size', y_test.shape[0])
```

#### c. Apply the Voting and Stacking methods on the dataset using models from (b) as the base estimators and evaluate the performance. You can use logistic regression, linear regression or gradient boosted trees as the meta learner for stacking.

```python
#from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier, StackingClassifier
```

```python
# Define base models
dt_model2 = DecisionTreeClassifier(random_state=100)
nb_model2 = GaussianNB()

# Train the base models
dt_model2.fit(X_train, y_train)
nb_model2.fit(X_train, y_train)
```

##### 1.Voting Ensemble

```python
# Voting ensemble
voting_ensemble = VotingClassifier(estimators=[('dt', dt_model2), ('nb', nb_model2)], voting='soft')
voting_ensemble.fit(X_train, y_train)
```

```python
#from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Predict and evaluate the voting ensemble
y_pred_voting = voting_ensemble.predict(X_test)
cm_voting = confusion_matrix(y_test, y_pred_voting)
accuracy_voting = accuracy_score(y_test, y_pred_voting)
precision_voting = precision_score(y_test, y_pred_voting, average='micro')
recall_voting = recall_score(y_test, y_pred_voting, average='micro')
f1_voting = f1_score(y_test, y_pred_voting, average='micro')

print('Voting Ensemble:')
print('Confusion Matrix:\n', cm_voting)
print('Accuracy: %.3f' % accuracy_voting)
print('Precision: %.3f' % precision_voting)
print('Recall: %.3f' % recall_voting)
print('F1-score: %.3f' % f1_voting)
```

##### 2.Stacking ensemble

```python
# Stacking ensemble
meta_learner = LogisticRegression()
stacking_ensemble = StackingClassifier(estimators=[('dt', dt_model2), ('nb', nb_model2)], final_estimator=meta_learner, cv=5)
stacking_ensemble.fit(X_train, y_train)
```

```python
#from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Predict and evaluate the stacking ensemble
y_pred_stacking = stacking_ensemble.predict(X_test)
cm_stacking = confusion_matrix(y_test, y_pred_stacking)
accuracy_stacking = accuracy_score(y_test, y_pred_stacking)
precision_stacking = precision_score(y_test, y_pred_stacking, average='micro')
recall_stacking = recall_score(y_test, y_pred_stacking, average='micro')
f1_stacking = f1_score(y_test, y_pred_stacking, average='micro')

print('\nStacking Ensemble:')
print('Confusion Matrix:\n', cm_stacking)
print('Accuracy: %.3f' % accuracy_stacking)
print('Precision: %.3f' % precision_stacking)
print('Recall: %.3f' % recall_stacking)
print('F1-score: %.3f' % f1_stacking)
```

#### d. Compare and comment on the performance between the Voting and Stacking ensembles, as well as the performance of the single models.

##### 1.Voting vs Stacking ensemble

When looking at the metrics (Confusion Matrix, Accuracy, Precision, Recall, F1-score) they both got the same results. Several factors could be contributing to this result:

1. Simple Dataset: iris is small & simple with only 150 samples & 4 features. Both methods may perform similarly as the dataset isn't complex enough for advantages of stacking over voting to be apparent (vice versa).
2. Base Models Performance: there's possibility that Decision Tree & Naive Bayes models might produce similar predictions. If the model perform similarly, the combined predictions from the voting classifier & stacking classifier might also end up being similar.
3. Base Models Performance: if the stacking classifier base models used have similar performance, the stacking method may not provide significant improvement over the voting method. As stacking typically benefits from diversity in base models.
4. Meta-Learner: the Logistic regression in stacking classifier might be producing similar results to voting method. Especially if the base models are not very diverse or if the Logistic regression model is not adding significant value.

In more complex datasets with more features and variability, the differences between voting and stacking method might become more pronounced.
