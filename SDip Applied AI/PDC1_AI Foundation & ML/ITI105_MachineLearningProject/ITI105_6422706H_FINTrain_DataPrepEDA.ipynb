*   ID - client loan application id.

*   approv_in_adv - indicates whether the loan was approved in advance (pre, nopre).
*   loan_type - type of loan (type1, type2, type3).
*   loan_purpose - purpose of the loan (p1, p2, p3, p4).
*   Credit_Worthiness - credit worthiness (l1, l2).
*   open_credit - indicates whether the applicant has any open credit accounts (opc, nopc).
*   business_or_commercial - indicates whether the loan is for business/commercial purposes (ob/c - business/commercial, nob/c - personal).
*   loan_amount - amount of money being borrowed.
*   rate_of_interest - interest rate charged on the loan.
*   Interest_rate_spread - difference between the interest rate on the loan and a benchmark interest rate.
*   Upfront_charges - initial charges associated with securing the loan.
*   term - duration of the loan in months.
*   Neg_ammortization - indicates whether the loan allows for negative ammortization (neg_amm, not_neg).
*   interest_only - indicates whether the loan has an interest-only payment option (int_only, not_int).
*   lump_sum_payment - indicates if a lump sum payment is required at the end of the loan term (lpsm, not_lpsm).
*   property_value - value of the property being financed.
*   construction_type - type of construction (sb - site built, mh - manufactured home).
*   occupancy_type - occupancy type (pr - primary residence, sr- secondary residence, ir - investment property).
*   Secured_by - specifies the type of collateral securing the loan (home, land).
*   total_units - number of units in the property being financed (1U, 2U, 3U, 4U).
*   income - applicant's annual income.
*   credit_type - applicant's type of credit (CIB - credit information bureau , CRIF - CIRF credit information bureau, EXP - experian , EQUI - equifax).

*   co-applicant_credit_type - co-applicant's type of credit (CIB - credit information bureau EXP - experian).

*   submission_of_application - indicates how the application was submitted (to_inst - to institution, not_inst - not to institution).

*   Region - geographic region where the property is located (North, south, central, North-East).
*   Security_Type - type of security or collateral backing the loan (direct, indirect).

Note: replace missing values to zero values, due to business context.


*   rate_of_interest : Has 25,504 missing values. Which is justifiable as loan or financial product can be 0%. It can occur for situations like e.g. promo offers, subsidised loans, zero-interest financing, no-interest savings accounts, negative/zero nominal rates.

*   Interest_rate_spread : Has 25,647 missing values. Which is justifiable as if the interest rate on the loan or investment is exactly equal to the benchmark interest rate, the spread would be 0. It can occur for situations where e.g. the loan/investment is offered at market rates without any additional margin, or the benchmark rate & the loan rate are effectively the same due to market conditions/ the structure of the product.

*   Upfront_charges : Has 27,756 missing values. Which is justifiable as it could mean there are no initial costs required to proceed with the transaction or service. It can occur for situations like e.g. promo offers, no initial fees, negotiation or special terms.

*   LTV : Has 10,579 missing values. Which is justifiable as an LTV ratio of 0% indicates that there is no loan amount relative to the property's value. It can occur for situations like e.g. no mortgage, full equity. It typically indicates that the property is fully owned outright, withuot any outstanding debt against it.

*   dtir1 : Has 16,841 missing values. Which is justifiaable as debt-to-income (DTI) ratio of 0% indicates that the individul has no monthly debt payments relative to their income. It generally suggests a very strong financial position in terms of debt management, as there are no debt payments impacting the individual's income.

### 2.3 Categorical Missing Value Imputation
Technique: Most frequent observation.

```python
# Impute values using 'most frequent observation' for columns:
# "loan_limit", "approv_in_adv", "loan_purpose", "Neg_ammortization", "age", "submission_of_application"

def fill_with_most_frequent(df, columns):
    """
    Fill missing values in specified columns with the most frequent value.

    Parameters:
    - df: pandas DataFrame
    - columns: list of columns to process
    """
    for column in columns:
        most_frequent = df[column].value_counts().idxmax()  # Get the most frequent value
        df[column].fillna(most_frequent, inplace=True)

# List of columns to fill with the most frequent value
categorical_columns = ["loan_limit", "approv_in_adv", "loan_purpose", "Neg_ammortization", "age", "submission_of_application"]

# Apply the function to fill missing values
fill_with_most_frequent(data, categorical_columns)
```

### 2.6 Detect & Remove Outliers
Technique: based on the Z-score method.

1. Z-Score Calculation:
*   For each numerical column in the DataFrame, the Z-score is calculated for each value.
*   The Z-score is defined as (x - mean)/std, where x is the value, and mean and std are the mean and standard deviation of the column, respectively.
*   LValues with a Z-score greater than the 'threshold' (usually 3) are considered outliers.


2. Filtering Outliers:
*   Rows where any numerical column has an absolute Z-score greater than the threshold are removed from the DataFrame.
*   Note for performance consideration:Calsulating Z-score and filtering can be optimized. It's better to use vectorised operations when working with large datasets.


```python
# Define a function

#def detect_outliers(data, threshold=3):
    # Create a boolean mask for the rows to keep
#    mask = np.ones(len(data), dtype=bool)

 #   for column in data.columns:
 #       if data[column].dtype in ['float64', 'int64']:
 #           mean = np.mean(data[column])
 #           std = np.std(data[column])
 #           z_scores = np.abs((data[column] - mean) / std)

            # Update the mask to exclude rows where z-score is greater than the threshold
 #           mask &= (z_scores <= threshold)

 #   return data[mask]
```

```python
# Use this Define a Function

def detect_outliers(data_clean, threshold=3):
    # Create a boolean mask for the rows to keep
    mask = np.ones(len(data_clean), dtype=bool)

    # Iterate over each numeric column
    for column in data_clean.select_dtypes(include=['float64', 'int64']).columns:
        mean = np.mean(data_clean[column])
        std = np.std(data_clean[column])
        z_scores = np.abs((data_clean[column] - mean) / std)

        # Update the mask to exclude rows where z-score is greater than the threshold
        mask &= (z_scores <= threshold)

    # Apply the mask to the DataFrame
    return data_clean[mask]
```

### 4.2 Categorical Correlation
Technique: V Cramer's
Disclaimer: Drop categorical columns with low correlation, if any.

*   It is based on a Pearson's Chi-Square Test.
*   Similarly to correlation, the output is in the range of [0,1]. Where 0 eans no association & 1 is full association.
*   There are no negative values, as there's no such thing as a negative association.



```python
# Filter Categorical Columns/Features
df_categorical = df.select_dtypes(include=['object'])
df_categorical
```

```python
from scipy.stats import chi2_contingency

# Building of the Cramer's V function
def cramers_v(var1, var2):
    crosstab = np.array(pd.crosstab(var1, var2, rownames=None, colnames=None)) # Cross table building
    stat = chi2_contingency(crosstab)[0] # Keeping of the test statistic of the Chi2 test
    obs = np.sum(crosstab) # Number of observations
    mini = min(crosstab.shape) - 1 # Take the minimum value between the columns and the rows of the cross table
    return (stat / (obs * mini))

```

```python
# Building of the matrix
rows = []
for var1 in df_categorical:
    col = []
    for var2 in df_categorical:
        cramers = cramers_v(df_categorical[var1], df_categorical[var2]) # Cramer's V test
        col.append(round(cramers, 2)) # Keeping of the rounded value of the Cramer's V
    rows.append(col)

```

```python
# Create DataFrame for Cramer's V results
cramers_results = np.array(rows)
vCramers = pd.DataFrame(cramers_results, columns=df_categorical.columns, index=df_categorical.columns)

# Display the Cramer's V matrix
print(vCramers)
```

```python
# Creating heatMap for the Cramer's V matrix
plt.figure(figsize=(10, 8))
sns.heatmap(vCramers, annot=True, cmap="YlGnBu")
plt.title('Cramér\'s V Heatmap')
plt.show()
```

### 4.5 Encoding for Categorical Columns (LabelEncoder & OneHot Encoder)

```python
from sklearn.preprocessing import LabelEncoder

# Dictionary to store label encoders for each column
label_encoders = {}

# List of columns to encode
columns_to_encode = ["Credit_Worthiness", "total_units"]

# Encode specified categorical columns and store the label encoders
for col in columns_to_encode:
    if col in df.columns and df[col].dtype == 'object':
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
        label_encoders[col] = le
```

```python
# Display the label encoding mapping for each column
for col, le in label_encoders.items():
    print(f"\nMapping for {col}:")
    print(dict(zip(le.classes_, le.transform(le.classes_))))
```

```python
# Define the desired order of categories
age_order = ['<25-34', '35-44', '45-54', '55-64', '65-74>']

# Create a mapping from the category to an integer
age_mapping = {category: index for index, category in enumerate(age_order)}

# Apply the mapping to the 'age' column
df['age'] = df['age'].map(age_mapping)
```
