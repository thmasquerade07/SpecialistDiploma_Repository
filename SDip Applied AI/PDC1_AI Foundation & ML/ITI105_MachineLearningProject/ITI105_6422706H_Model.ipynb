```python
def plot_cap_curve(model, X_test, y_test, model_name):
    # Obtain the model's predictions
    if hasattr(model, "predict_proba"):
        y_pred_prob = model.predict_proba(X_test)[:, 1]
    else:
        # For SVM, use decision_function instead of predict_proba
        y_pred_prob = model.decision_function(X_test)

    # Sort predictions by their predicted probability in descending order
    y_test_sorted = y_test[np.argsort(y_pred_prob)[::-1]]

    # Calculate the cumulative sum of positives and total positives
    cum_true_positives = np.cumsum(y_test_sorted == 1.0)
    total_positives = np.sum(y_test_sorted == 1.0)

    # Calculate the CAP curve values
    cap_values = cum_true_positives / total_positives

    # Calculate the x-axis points (percentile of samples)
    cap_x = np.arange(1, len(cap_values) + 1) / len(cap_values)

    # Plot the CAP curve
    plt.plot(cap_x, cap_values, label=f'{model_name}')
```

```python
# Initialize figure for CAP curves
plt.figure(figsize=(10, 8))

# Plot CAP curves for each model
models = [
    ('Gaussian Naive Bayes', gnb),
    ('Complement Naive Bayes', cnb),
    ('Random Forest', rf),
    ('SVM RBF', svm_rbf)
]

for model_name, model in models:
    plot_cap_curve(model, X_test, y_test, model_name)

# Plot the random performance line (outside the loop)
plt.plot([0, 1], [0, 1], 'k--', label='Random Model')

# Plot the perfect performance line (outside the loop)
#perfect_performance_x = np.linspace(0, 1, 100)
#perfect_performance_y = np.linspace(0, 1, 100)
#plt.plot(perfect_performance_x, perfect_performance_y, color='grey', linestyle='-', label='Perfect Model')

# Set plot title and labels
plt.title('Cumulative Accuracy Profile (CAP) Curve Comparison', fontsize=15, fontweight='bold')
plt.xlabel('Proportion of Total Samples', fontsize=12)
plt.ylabel('Proportion of Total Positive Samples', fontsize=12)

# Show legend
plt.legend(loc='lower right', fontsize=12)

# Show the plot
plt.show()
```

```python
# Filter Categorical Columns/Features
df_categorical = df.select_dtypes(include=['object'])

from scipy.stats import chi2_contingency

# Building of the Cramer's V function
def cramers_v(var1, var2):
    crosstab = np.array(pd.crosstab(var1, var2, rownames=None, colnames=None)) # Cross table building
    stat = chi2_contingency(crosstab)[0] # Keeping of the test statistic of the Chi2 test
    obs = np.sum(crosstab) # Number of observations
    mini = min(crosstab.shape) - 1 # Take the minimum value between the columns and the rows of the cross table
    return (stat / (obs * mini))

# Building of the matrix
rows = []
for var1 in df_categorical:
    col = []
    for var2 in df_categorical:
        cramers = cramers_v(df_categorical[var1], df_categorical[var2]) # Cramer's V test
        col.append(round(cramers, 2)) # Keeping of the rounded value of the Cramer's V
    rows.append(col)

# Create DataFrame for Cramer's V results
cramers_results = np.array(rows)
vCramers = pd.DataFrame(cramers_results, columns=df_categorical.columns, index=df_categorical.columns)

# Display the Cramer's V matrix
#print(vCramers)

# Creating heatMap for the Cramer's V matrix
plt.figure(figsize=(10, 8))
sns.heatmap(vCramers, annot=True, cmap="YlGnBu")
plt.title('Cramér\'s V Heatmap')
plt.show()
```

```python
def plot_cap_curve(model, X_test, y_test, model_name):
    # Obtain the model's predictions
    if hasattr(model, "predict_proba"):
        y_pred_prob = model.predict_proba(X_test)[:, 1]
    else:
        # For SVM, use decision_function instead of predict_proba
        y_pred_prob = model.decision_function(X_test)

    # Sort predictions by their predicted probability in descending order
    y_test_sorted = y_test[np.argsort(y_pred_prob)[::-1]]

    # Calculate the cumulative sum of positives and total positives
    cum_true_positives = np.cumsum(y_test_sorted == 1.0)
    total_positives = np.sum(y_test_sorted == 1.0)

    # Calculate the CAP curve values
    cap_values = cum_true_positives / total_positives

    # Calculate the x-axis points (percentile of samples)
    cap_x = np.arange(1, len(cap_values) + 1) / len(cap_values)

    # Plot the CAP curve
    plt.plot(cap_x, cap_values, label=f'{model_name}')
```

```python
# Initialize figure for CAP curves
plt.figure(figsize=(10, 8))

# Plot CAP curves for each model
models = [
#    ('Gaussian Naive Bayes', gnb),
    ('Complement Naive Bayes', cnb),
    ('Random Forest', rf),
    ('SVM RBF', svm_rbf)
]

for model_name, model in models:
    plot_cap_curve(model, X_test, y_test, model_name)

# Plot the random performance line (outside the loop)
plt.plot([0, 1], [0, 1], 'k--', label='Random Model')

# Plot the perfect performance line (outside the loop)
#perfect_performance_x = np.linspace(0, 1, 100)
#perfect_performance_y = np.linspace(0, 1, 100)
#plt.plot(perfect_performance_x, perfect_performance_y, color='grey', linestyle='-', label='Perfect Model')

# Set plot title and labels
plt.title('Cumulative Accuracy Profile (CAP) Curve Comparison', fontsize=15, fontweight='bold')
plt.xlabel('Proportion of Total Samples', fontsize=12)
plt.ylabel('Proportion of Total Positive Samples', fontsize=12)

# Show legend
plt.legend(loc='lower right', fontsize=12)

# Show the plot
plt.show()
```

```python
def plot_cap_curve(model, X_test, y_test, model_name):
    # Obtain the model's predictions
    if hasattr(model, "predict_proba"):
        y_pred_prob = model.predict_proba(X_test)[:, 1]
    else:
        # For SVM, use decision_function instead of predict_proba
        y_pred_prob = model.decision_function(X_test)

    # Sort predictions by their predicted probability in descending order
    y_test_sorted = y_test[np.argsort(y_pred_prob)[::-1]]

    # Calculate the cumulative sum of positives and total positives
    cum_true_positives = np.cumsum(y_test_sorted == 1.0)
    total_positives = np.sum(y_test_sorted == 1.0)

    # Calculate the CAP curve values
    cap_values = cum_true_positives / total_positives

    # Calculate the x-axis points (percentile of samples)
    cap_x = np.arange(1, len(cap_values) + 1) / len(cap_values)

    # Plot the CAP curve
    plt.plot(cap_x, cap_values, label=f'{model_name}')
```

```python
# Initialize figure for CAP curves
plt.figure(figsize=(10, 8))

# Plot CAP curves for each model
models = [
#    ('Gaussian Naive Bayes', gnb),
    ('Complement Naive Bayes', cnb),
    ('Random Forest', rf),
#    ('SVM RBF', svm_rbf)
]

for model_name, model in models:
    plot_cap_curve(model, X_test, y_test, model_name)

# Plot the random performance line (outside the loop)
plt.plot([0, 1], [0, 1], 'k--', label='Random Model')

# Plot the perfect performance line (outside the loop)
#perfect_performance_x = np.linspace(0, 1, 100)
#perfect_performance_y = np.linspace(0, 1, 100)
#plt.plot(perfect_performance_x, perfect_performance_y, color='grey', linestyle='-', label='Perfect Model')

# Set plot title and labels
plt.title('Cumulative Accuracy Profile (CAP) Curve Comparison', fontsize=15, fontweight='bold')
plt.xlabel('Proportion of Total Samples', fontsize=12)
plt.ylabel('Proportion of Total Positive Samples', fontsize=12)

# Show legend
plt.legend(loc='lower right', fontsize=12)

# Show the plot
plt.show()
```

```python
def plot_cap_curve(model, X_test, y_test, model_name):
    # Obtain the model's predictions
    if hasattr(model, "predict_proba"):
        y_pred_prob = model.predict_proba(X_test)[:, 1]
    else:
        # For SVM, use decision_function instead of predict_proba
        y_pred_prob = model.decision_function(X_test)

    # Sort predictions by their predicted probability in descending order
    y_test_sorted = y_test[np.argsort(y_pred_prob)[::-1]]

    # Calculate the cumulative sum of positives and total positives
    cum_true_positives = np.cumsum(y_test_sorted == 1.0)
    total_positives = np.sum(y_test_sorted == 1.0)

    # Calculate the CAP curve values
    cap_values = cum_true_positives / total_positives

    # Calculate the x-axis points (percentile of samples)
    cap_x = np.arange(1, len(cap_values) + 1) / len(cap_values)

    # Plot the CAP curve
    plt.plot(cap_x, cap_values, label=f'{model_name}')
```

```python
# Initialize figure for CAP curves
plt.figure(figsize=(10, 8))

# Plot CAP curves for each model
models = [
    ('Gaussian Naive Bayes', gnb),
    ('Complement Naive Bayes', cnb),
    ('Random Forest', rf),
    ('SVM RBF', svm_rbf)
]

for model_name, model in models:
    plot_cap_curve(model, X_test, y_test, model_name)

# Plot the random performance line (outside the loop)
plt.plot([0, 1], [0, 1], 'k--', label='Random Model')

# Plot the perfect performance line (outside the loop)
#perfect_performance_x = np.linspace(0, 1, 100)
#perfect_performance_y = np.linspace(0, 1, 100)
#plt.plot(perfect_performance_x, perfect_performance_y, color='grey', linestyle='-', label='Perfect Model')

# Set plot title and labels
plt.title('Cumulative Accuracy Profile (CAP) Curve Comparison', fontsize=15, fontweight='bold')
plt.xlabel('Proportion of Total Samples', fontsize=12)
plt.ylabel('Proportion of Total Positive Samples', fontsize=12)

# Show legend
plt.legend(loc='lower right', fontsize=12)

# Show the plot
plt.show()
```
