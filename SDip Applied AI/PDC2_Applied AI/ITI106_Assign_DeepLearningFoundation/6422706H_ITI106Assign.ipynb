### 3.FNN model & tuning

#### 3.1 Number of layers and neuron unit

Layers & Neurons: We are testing three different strategies with a base activation function of RELU for all hidden layers:

1. 1 hidden layer with 8 neurons
2. 2 hidden layers with 16 and 8 neurons
3. 3 hidden layers with 32, 16, and 8 neurons

```python
from keras.models import Sequential
from keras.layers import Dense

def create_custom_model(input_dim, output_dim, hidden_layers, activations, name='model'):
    """
    Creates a customizable neural network model.

    Args:
        input_dim (int): Number of input features.
        output_dim (int): Number of output classes.
        hidden_layers (list of int): List specifying the number of nodes in each hidden layer.
        activations (list of str): Activation functions for each hidden layer.
        name (str): Name of the model.

    Returns:
        Sequential: Compiled Keras model.
    """
    model = Sequential(name=name)

    # Input layer
    model.add(Dense(hidden_layers[0], input_dim=input_dim, activation=activations[0]))

    # Hidden layers
    for nodes, activation in zip(hidden_layers[1:], activations[1:]):
        model.add(Dense(nodes, activation=activation))

    # Output layer
    model.add(Dense(output_dim, activation='softmax'))

    # Compile the model
    model.compile(loss='categorical_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])
    return model

# Model configurations
hidden_layer_configs = [
    [8],          # 1 hidden layer with 8 neurons
    [16, 8],      # 2 hidden layers with 16 and 8 neurons
    [32, 16, 8]   # 3 hidden layers with 32, 16, and 8 neurons
]
activation_configs = [
    ['relu'],              # 1 hidden layer activation
    ['relu', 'relu'],      # 2 hidden layers activations
    ['relu', 'relu', 'relu']  # 3 hidden layers activations
]

# Create models
models = [
    create_custom_model(input_dim=X_train.shape[1],
                        output_dim=y_train.shape[1],
                        hidden_layers=hidden_layers,
                        activations=activations,
                        name=f'model_{i+1}')
    for i, (hidden_layers, activations) in enumerate(zip(hidden_layer_configs, activation_configs))
]

# Display summaries
for model in models:
    model.summary()
```

```python
models = [
    lambda input_dim=X_train.shape[1], output_dim=y_train.shape[1], i=i: create_custom_model(
        input_dim=input_dim,
        output_dim=output_dim,
        hidden_layers=hidden_layers,
        activations=activations,
        name=f'model_{i+1}'  # Capture the current value of `i`
    )
    for i, (hidden_layers, activations) in enumerate(zip(hidden_layer_configs, activation_configs))
]
```

```python
from keras.callbacks import TensorBoard

# Dictionary to store training history and models
history_dict = {}

# TensorBoard Callback
tensorboard_callback = TensorBoard(log_dir="logs", histogram_freq=1)

# Train and evaluate models
for i, create_model in enumerate(models):
    # Create the model (call the function stored in models)
    model = create_model(input_dim=X_train.shape[1], output_dim=y_train.shape[1])
    print(f'Model name: {model.name}')

    # Train the model
    history_callback = model.fit(
        X_train, y_train,
        batch_size=5,
        epochs=50,
        verbose=0,
        validation_data=(X_test, y_test),
        callbacks=[tensorboard_callback]
    )

    # Evaluate the model
    score = model.evaluate(X_test, y_test, verbose=0)
    print(f'Test loss: {score[0]:.4f}')
    print(f'Test accuracy: {score[1]:.4f}')

    # Store the history and model
    history_dict[model.name] = {'history': history_callback, 'model': model}
```

```python
# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# Loop through each model's history in history_dict
for model_name, data in history_dict.items():
    history = data['history'].history  # Access the training history
    val_accuracy = history['val_accuracy']
    val_loss = history['val_loss']

    # Plot validation accuracy
    ax1.plot(val_accuracy, label=model_name, linewidth=2)

    # Plot validation loss
    ax2.plot(val_loss, label=model_name, linewidth=2)

# Customize accuracy subplot
ax1.set_title('Validation Accuracy Over Epochs', fontsize=14)
ax1.set_ylabel('Validation Accuracy', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.legend(title='Models', loc='lower right', fontsize=10)

# Customize loss subplot
ax2.set_title('Validation Loss Over Epochs', fontsize=14)
ax2.set_xlabel('Epochs', fontsize=12)
ax2.set_ylabel('Validation Loss', fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.7)
ax2.legend(title='Models', loc='upper right', fontsize=10)

# Tight layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

#### 3.2 Regularization

Regularization: We are testing the effects of using L2 regularization

```python
#from keras.models import Sequential
#from keras.layers import Dense
from keras.regularizers import l2
#from keras.callbacks import TensorBoard

# Model configurations
hidden_layer_configs = [
    [8],          # 1 hidden layer with 8 neurons
    [16, 8],      # 2 hidden layers with 16 and 8 neurons
    [32, 16, 8]   # 3 hidden layers with 32, 16, and 8 neurons
]

# Regularization strength
l2_regularizer = 0.01  # L2 regularization strength

def create_custom_model(input_dim, output_dim, hidden_layers, activations, l2_strength=0.01, name='model'):
    model = Sequential(name=name)

    # Add hidden layers with L2 regularization
    for i, units in enumerate(hidden_layers):
        if i == 0:
            model.add(Dense(units, input_dim=input_dim, activation=activations[i], kernel_regularizer=l2(l2_strength)))
        else:
            model.add(Dense(units, activation=activations[i], kernel_regularizer=l2(l2_strength)))

    # Add output layer
    model.add(Dense(output_dim, activation='softmax'))

    # Compile model
    model.compile(loss='categorical_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])

    return model

# Generate models with the new configurations
models = [
    lambda input_dim=X_train.shape[1], output_dim=y_train.shape[1], i=i: create_custom_model(
        input_dim=input_dim,
        output_dim=output_dim,
        hidden_layers=hidden_layers,
        activations=['relu'] * len(hidden_layers),  # Apply ReLU activation to each hidden layer
        l2_strength=l2_regularizer,
        name=f'model_{i+1}'  # Capture the current value of `i`
    )
    for i, hidden_layers in enumerate(hidden_layer_configs)
]
```

```python
# Dictionary to store training history and models
history_dict = {}

# TensorBoard Callback
tensorboard_callback = TensorBoard(log_dir="logs", histogram_freq=1)

# Train and evaluate models
for i, create_model in enumerate(models):
    # Create the model (call the function stored in models)
    model = create_model(input_dim=X_train.shape[1], output_dim=y_train.shape[1])
    print(f'Model name: {model.name}')

    # Train the model
    history_callback = model.fit(
        X_train, y_train,
        batch_size=5,
        epochs=50,
        verbose=0,
        validation_data=(X_test, y_test),
        callbacks=[tensorboard_callback]
    )

    # Evaluate the model
    score = model.evaluate(X_test, y_test, verbose=0)
    print(f'Test loss: {score[0]:.4f}')
    print(f'Test accuracy: {score[1]:.4f}')

    # Store the history and model
    history_dict[model.name] = {'history': history_callback, 'model': model}
```

```python
# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# Loop through each model's history in history_dict
for model_name, data in history_dict.items():
    history = data['history'].history  # Access the training history
    val_accuracy = history['val_accuracy']
    val_loss = history['val_loss']

    # Plot validation accuracy
    ax1.plot(val_accuracy, label=model_name, linewidth=2)

    # Plot validation loss
    ax2.plot(val_loss, label=model_name, linewidth=2)

# Customize accuracy subplot
ax1.set_title('Validation Accuracy Over Epochs', fontsize=14)
ax1.set_ylabel('Validation Accuracy', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.legend(title='Models', loc='lower right', fontsize=10)

# Customize loss subplot
ax2.set_title('Validation Loss Over Epochs', fontsize=14)
ax2.set_xlabel('Epochs', fontsize=12)
ax2.set_ylabel('Validation Loss', fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.7)
ax2.legend(title='Models', loc='upper right', fontsize=10)

# Tight layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

#### 3.5 Initialization of weights

Weight Initializers: We are testing three different weight initialization strategies:

1. he_normal: A good initializer for ReLU activation
(normal distribution with mean 0 and variance based on the number of inputs).
2. glorot_uniform: A uniform distribution with values chosen according to the size of the layer.
3. random_normal: A random normal distribution, with a mean of 0 and a standard deviation of 0.05 (can adjust the standard deviation based on preferences).

```python
#from keras.models import Sequential
#from keras.layers import Dense
from keras.initializers import HeNormal, GlorotUniform, RandomNormal
#from keras.callbacks import TensorBoard

# Model configurations
hidden_layer_configs = [
    [8],          # 1 hidden layer with 8 neurons
    [16, 8],      # 2 hidden layers with 16 and 8 neurons
    [32, 16, 8]   # 3 hidden layers with 32, 16, and 8 neurons
]

# Weight initializers to test
initializers = ['he_normal', 'glorot_uniform', 'random_normal']

def create_custom_model(input_dim, output_dim, hidden_layers, activation='relu', optimizer='adam', initializer='he_normal', name='model'):
    model = Sequential(name=name)

    # Add hidden layers with specified activations and weight initializer
    for i, units in enumerate(hidden_layers):
        if i == 0:
            model.add(Dense(units, input_dim=input_dim, activation=activation, kernel_initializer=initializer))
        else:
            model.add(Dense(units, activation=activation, kernel_initializer=initializer))

    # Add output layer
    model.add(Dense(output_dim, activation='softmax'))

    # Compile model with the specified optimizer
    model.compile(loss='categorical_crossentropy',
                  optimizer=optimizer,
                  metrics=['accuracy'])

    return model
```

```python
# Dictionary to store training history and models
history_dict = {}

# TensorBoard Callback
tensorboard_callback = TensorBoard(log_dir="logs", histogram_freq=1)

# Test different initializers with Adam optimizer
for initializer in initializers:
    print(f'\nTesting models with initializer: {initializer}')

    # Generate models with the current weight initializer
    models = [
        lambda input_dim=X_train.shape[1], output_dim=y_train.shape[1], i=i, initializer=initializer: create_custom_model(
            input_dim=input_dim,
            output_dim=output_dim,
            hidden_layers=hidden_layers,
            activation='relu',  # ReLU activation as requested
            optimizer='adam',  # Using Adam optimizer
            initializer=initializer,
            name=f'model_{i+1}_{initializer}'  # Include the initializer in the model name
        )
        for i, hidden_layers in enumerate(hidden_layer_configs)
    ]

    # Train and evaluate models
    for i, create_model in enumerate(models):
        # Create the model (call the function stored in models)
        model = create_model(input_dim=X_train.shape[1], output_dim=y_train.shape[1])
        print(f'Model name: {model.name}')

        # Show model summary
        model.summary()

        # Train the model
        history_callback = model.fit(
            X_train, y_train,
            batch_size=5,
            epochs=50,
            verbose=0,
            validation_data=(X_test, y_test),
            callbacks=[tensorboard_callback]
        )

        # Evaluate the model
        score = model.evaluate(X_test, y_test, verbose=0)
        print(f'Test loss: {score[0]:.4f}')
        print(f'Test accuracy: {score[1]:.4f}')

        # Store the history and model
        history_dict[model.name] = {'history': history_callback, 'model': model}
```

### 4.Discussion: FNN vs CNN vs RNN

#### 4.1 FNN vs CNN

When solving a classification problem using the Iris dataset, FNN & CNN will perform differently based on the nature of the problem & the data. Below are some key differences between these two types of deep learning architectures:

1. Network Structure

>*   FNN: The Iris dataset has 4 continuous features which makes it a suitable architecture for classification in this case. It is ideal for problems where the relationships between features are complex.
*   CNN: It is primarily designed for image-related tasks where there is a strong spatial or temporal relationship between input datapoints (e.g. images or videos). Since the dataset consists of tabular data with no inherent spatial relationships between features, it will not be an optimal choice.



2. Data Handling

>*   FNN: Accepts input as flat vectors, e.g. the 4 features of the Iris dataset. For tabular datasets like Iris, preprocessing often includes scaling the data, handling missing values, and encoding categorical variables (if any).
*   CNN: Expects input data in the form of multidimensional arrays e.g. 2D images or 3D video data. It often requires preprocessing of image data to be normalized & augmented to improve generalization & prevent overfitting (e.g. rotations, translations).

3. Feature Extraction

>*   FNN: The network learns global relationships between features in a fully connected manner. It does not explicitly perform feature extration like CNNs. Instead, it relies on the hidden layers to learn these relationships.
*   CNN: The network are specifically designed to learn hierarchical feature representations through convolutional layers. These layers automatically detect local patterns (e.g. edges in images) in the early layers & combine them into more complex patterns in the deeper layers.

In conclusion, FNN is a more suited for classification problems with tabular data format like Iris dataset. There is no inherent spatial structure between the features for a CNN to ingest. If an image of the Iris samples was given, then it would provide significant advantages since it will be more appropriate as spatial relationship patterns can be learned & exploited.

#### 4.2 FNN vs RNN

When solving a classification problem using the Iris dataset, both FNN & RNN are potential choices, but they are suited to different types of data. Below are some key differences between these two types of deep learning architectures:

1. Network Structure

>*   FNN: The datapoints are independent of each other, making it suitable for tabular data or classification problems where each sample is treated as a separate instance, without considering the order or sequence of the datapoints. Thus it can learn the relationhips between the features & perform well in classification tasks.
*   RNN: Are designed to handle sequential data with its recurrent connection capabilities to learn temporal dependencies & sequencial relationships. It is best suited for tasks where there's an inherent temporal or sequential dependency between data points (e.g. time-series forcasting, natural language processing - text or speech, video analysis).

2. Data Handling

>*   FNN: Accepts data in the form of a fixed-size vector for each instance. (i.e. the Iris dataset features are simply a 4D vector for each sample). Works well if preprocessed tabular features like feature scaling & encoding of categorical data (if applicable).
*   RNN:Expects sequential data where each input datapoint is part of a sequence. This can be time-series data or sequences of token in text, where the output of each step depends on previous ones. Often, its data requirements may involve passing or reshaping for ficed-length sequences. Since Iris dataset has no temporal or sequential dependency between features, it is unnecessary & inefficient for such use case.

3. Feature Learning

>*   FNN: The network learns globa relationships between input features with no regards to any sequence or order. It performs well for tasks where each input is independent of others, such as classification on tabular datasets like Iris. It does this by learning weights in each layer during training. Hidden layers are used to map the raw input features into higher-level representations that are then classified into categories.
*   RNN: The network excels at learning temporal dependencies & sequential patterns in data. They are designed to process data step-by-step while retaining information from previous time steps in their hidden states, thus allows to capture long-term dependencies in sequential data. This ability will be wasted on no sequential data like Iris.

In conclusion, FNN is a more appropriate choice as it can quickly & accurately learn the relationships between the features & the target classes, making them ideal for this classification task. Since the Iris dataset consists of tabular data with no inherent sequence or time dependency, it is unnecessary for this use case.
